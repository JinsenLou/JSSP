````python
class JSSP_Instance:
    
    def __init__(self):
        self.name = ''
        self.jobs = 0
        self.machines = 0
        self.job_data = []
        
    def read_instance(self, instance_name, filename):
        #Get contents
        with open(filename, 'r') as x:
            CONTENTS = x.readlines()

        self.name = instance_name

        for i,data in enumerate(CONTENTS):
            data = data.replace('\n','')
            #Get data with respective instance name

            if instance_name in data:
                plus = 0
                while plus < 2:
                    plus += ('+' in CONTENTS[i])
                    _split = CONTENTS[i].split()
                    i += 1
                    try:
                        _split = [int(num) for num in _split]
                    except:
                        continue
                    #get numbers
                    if len(_split) == 2:
                        self.machines, self.jobs = _split[1], _split[0]
                    elif len(_split) > 2:
                        self.job_data = self.job_data + [_split]



    def __str__(self):
        # Build the string representation of the object
        res = []
        res.append(f"Instance: {self.name}")
        res.append(f"Jobs: {self.jobs}")
        res.append(f"Machines: {self.machines}")
        res.append("Jobs:")
        for i in range(self.jobs):
            job_str = f"J{i}\t["
            for j in range(0,self.machines *2,2):
                job_str += f" {self.job_data[i][j]} {self.job_data[i][j+1]} "
            job_str += "]"
            res.append(job_str)

        return "\n".join(res)
````

````python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import random

class JSSP_Candidate_solution:
    def __init__(self, schedule):
        #schedule refers to (Machine, job) -> where job = [job1, time1, time2, job2 ....]
        self.schedule = schedule
        self.uniquejobs = []
        self.jobs = {} #jobs and their schedule
        self.uniqueJobs()
        
        
    def uniqueJobs(self):
        #gets unique jobs, no duplicates
        for machine in self.schedule:
            for i in range(0,len(machine),3):
                job = machine[i] #get the job
                self.uniquejobs.append(job)
        self.uniquejobs = list(set(self.uniquejobs))
        
    def gantt(self):
        #plots gantt charts
        colors = list(mcolors.CSS4_COLORS.keys())
        random.shuffle(colors)
        job_label = []

        fig, ax = plt.subplots(figsize=(10, 5))
        for i in range(len(self.schedule)):
            job_schedule = self.schedule[i]
            for j in range(0, len(job_schedule), 3):

                if job_schedule[j] not in job_label:
                    the_label = f'Job {job_schedule[j]}'
                    job_label.append(job_schedule[j])
                else:
                    the_label = None

                start_time = job_schedule[j + 1]
                duration = job_schedule[j + 2] - start_time

                ax.barh(f"M{i}", duration, left=start_time, height=0.5, align='center',
                        label=the_label, color=colors[job_schedule[j]])

        ax.set_xlabel('Time')
        ax.set_yticks([f"M{i}" for i in range(len(self.schedule))])
        ax.set_title('Gantt Chart')
        ax.legend(title='Job', loc='upper right', bbox_to_anchor=(1.05, 1), borderaxespad=0.)
        plt.show()

    def __str__(self):
        result = ""
        for idx, machine in enumerate(self.schedule):
            result += f"M{idx} [{', '.join(map(str, machine))}]\n"
        return result.rstrip()

    def consistent(self):
        #initiate with empty jobs
        #no overlap checks
        emptylist = [[] for i in range(len(self.uniquejobs))]
        self.jobs = dict(zip(self.uniquejobs, emptylist))
        jobs = self.jobs
        for machine in self.schedule:
            for i in range(0,len(machine),3):
                jobs[machine[i]].append([machine[i+1],machine[i+2]])

        for keys,values in jobs.items():
            sorted_values = sorted(values, key=lambda x: x[0])
            for i in range(len(sorted_values)-1):
                if sorted_values[i][1] > sorted_values[i+1][0]:
                    return False

        for machine in self.schedule:
            for i in range(0,len(machine)-3,3):
                if machine[i+2] > machine[i+4]:
                    return False
        return True
    
    def satisfies(self,instance):
        isconsistent = self.consistent()
        for i,job in enumerate(instance.job_data):
            job_time = [job[j+1] for j in range(0,len(job),2)]
            job_time = sorted(job_time)
            job_time_candidate = [k[1]-k[0] for k in self.jobs[i]]
            job_time_candidate = sorted(job_time)
            if job_time != job_time_candidate:
                return False
        return isconsistent
    
    def makespan(self):
        return max([sublist[-1] for sublist in self.schedule])
        
````
